### Huddersfield Town 2017-18 VS 2018-19 Comparison

#### Context
Huddersfield Town did not quite have the highs of Sheffield United or Leeds United in their first season after promotion from the EFL Championship but it was still a huge success. Highly fancied to finish last, they achieved a 16th place finish, relegating previousy established Premier League clubs such as West Brom and Stoke City in the process. However, they had an unmitigated disaster of a second season. They achieved a measly 16 points and were relegated in March becoming one of only a select group to do so. Here we will try to better understand the numbers behind Huddersfield's two seasons in the PL.

```python
# first let's investigate squad turnover
fig, (ax1, ax2) = plt.subplots(1,2, figsize = (20, 8))
df1.plot.barh(x = 'Player', y = 'Min', ax = ax1, title = 'Number of minutes per player 2017-18',
             color = 'tan')
df2.plot.barh(x = 'Player', y = 'Min', ax = ax2, title = 'Number of minutes per player 2018-19',
             color = 'cornflowerblue')
plt.show()
```

```python
print("The number of players in Huddersfield Town's 2017-18 squad was: " + str(len(df1.index)))
print("The number of players in Huddersfield Town's 2018-19 squad was: " + str(len(df2.index)))
```

```python
print("Mean minutes played in Huddersfield Town's 2017-18 squad was: " + str(df1['Min'].mean()))
print("Mean minutes played in Huddersfield Town's 2018-19 squad was: " + str(df2['Min'].mean()))
```

Huddersfield Town used far more players and there was a noticeable reduction in the average playing time of their squad (around 300 mins so over 3 games worth) which is a common trend amongst teams that struggled in their second season in the PL. It is a cliche in football about having a settled team but it appears to hold true to some extent here.

##### Let's look at performance metrics now. Firstly, we will focus on expected goals.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))
df_c1 = df1.sort_values('xG')
df_c2 = df2.sort_values('xG')
df_c1.plot.barh(x = 'Player', y = 'xG', ax = ax1, title = 'xG per player 2017-18',
             color = 'tan')
df_c2.plot.barh(x = 'Player', y = 'xG', ax = ax2, title = 'xG per player 2018-19',
             color = 'cornflowerblue')
plt.show()
```

```python
print("Team xG in Huddersfield Town's 2017-18 squad was: " + str(sum(df1['xG'])))
print("Team xG in Huddersfield Town's 2018-19 squad was: " + str(sum(df2['xG'])))
```

These numbers are very low for both seasons which suggests one of two things:
- Huddersfield Town were a team that relied heavily on a solid defence or
- they really struggled to break down Premier League defences.

#### xG VS Actual Difference

### Attacking numbers takeaways
Huddersfield Town didn't really have anyone across these two seasons who massively overperformed their xG. The highest overperformance was in their first season in which two of their attackers Steve Mounie and Laurent Depoitre had an overperformance of around 2. The rest either were roughly in line or had a reasonable underperformance. Mounie actually had an underperformance of around 4 the following season whereas in their first season Tom Ince had over 4 in underperformance which is quite a large number in this context. These numbers point towards a team that lacked the individual quality to put away chances, which leans us toward a lack of attacking quality as a reason for Huddersfield's weak second season and relegation.

### Chance creation numbers
The underlying numbers across both seasons show little surprise for Huddersfield's eventual relegation. They were only expected to assist approximately 24 and 20 goals in the respective seasons, which are incredibly low numbers. Even more worringly in both seasons they underperformed these numbers to quite a large extent (roughly 5 assists for the first season and roughly 7 for the second). This further points towards a team devoid of any real attacking quality and reliant on a solid defence and/or set pieces/mistakes from the opposition to score goals. Given the low numbers here it is hardly surprising Huddersfield Town fell foul to "second season syndrome".

### Per 90 comparisons

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (20,8))
df_c1 = df1.sort_values('Per90npxG+xAG')
df_c2 = df2.sort_values('Per90npxG+xAG')
df_c1.plot.barh(x = 'Player', y = 'Per90npxG+xAG', ax = ax1, title = 'Per90npxG+xAG per player 2017-18',
             color = 'tan')
df_c2.plot.barh(x = 'Player', y = 'Per90npxG+xAG', ax = ax2, title = 'Per90npxG+xAG per player 2018-19',
             color = 'cornflowerblue')
plt.show()
```

### Let's now compare players who played in both seasons for Huddersfield Town
By doing this, it allows us to better understand if for example a significant turnover of players or a loss of form from previously key players led to their "second season syndrome". We will focus on minutes played and per 90 metrics here.

```python
players1 = list(df1['Player'])
players2 = list(df2['Player'])
d = {}
for i in range(len(players1)):
    if players1[i] in players2:
        val = df2.loc[df2['Player'] == players1[i], "Per90npxG+xAG"].item()
        vals = val - df1.loc[df1['Player'] == players1[i], "Per90npxG+xAG"].item()
        d[players1[i]] = vals
    else:
        continue
```

```python
per90commonplayers = pd.DataFrame.from_dict(d, orient='index')
```

```python
per90commonplayers = per90commonplayers.reset_index()
per90commonplayers.rename(columns = {'index': 'Name', 0: 'SeasonDifferenceXGXA'}, inplace = True)
```

```python
players1 = list(df1['Player'])
players2 = list(df2['Player'])
d1 = {}
for i in range(len(players1)):
    if players1[i] in players2:
        val = df2.loc[df2['Player'] == players1[i], "Min"].item()
        vals = val - df1.loc[df1['Player'] == players1[i], "Min"].item()
        d1[players1[i]] = vals
    else:
        continue
```

```python
minscommonplayers = pd.DataFrame.from_dict(d1, orient = 'index')
minscommonplayers = minscommonplayers.reset_index()
minscommonplayers.rename(columns = {'index': 'Name', 0: 'SeasonDifferenceMins'}, inplace = True)
```

```python
result = pd.merge(minscommonplayers, per90commonplayers, on = ["Name"])
```

```python
result.head()
```

```python
result.plot.bar(x = 'Name', y = 'SeasonDifferenceXGXA', rot = 40, figsize = (18, 6),
                title = "Per90XG+XA Difference for Huddersfield Town's players who played in 2017/18 and 2018/19 seasons",
                ylabel = 'Per90XG+XA', xlabel = 'Player Name',
                color = (result['SeasonDifferenceXGXA'] > 0).map({True: 'g', False: 'r'}))
```

```python
result.plot.bar(x = 'Name', y = "SeasonDifferenceMins", rot = 40, figsize = (18, 6),
                title = "Minutes Played Difference for Huddersfield Town's players who played in 2017/18 and 2018/19 seasons",
                ylabel = 'Minutes Played Difference', xlabel = 'Player Name',
                color = (result['SeasonDifferenceMins'] > 0).map({True: 'limegreen', False: 'red'}))
fig.patch.set_facecolor('white')
```

### Goalkeeper Analysis

```python
gkseason1 = pd.read_csv("datasets/cleanedAdvanced/huddersfield-adkeeper1718-cleaned.csv")
gkseason2 = pd.read_csv("datasets/cleanedAdvanced/huddersfield-adkeeper1819-cleaned.csv")
```

```python
gkseason1.head()
```

```python
gkseason1.columns
```

```python
keepers1 = list(gkseason1['Player'])
keepers2 = list(gkseason2['Player'])
keeperdict1 = {}
keeperdict2 = {}
for i in range(len(keepers1)):
    if keepers1[i] in keepers2:
        keeperdict1[keepers1[i]] = gkseason1.loc[gkseason1['Player'] == keepers1[i], "ExpectedPSxG+/-"].item()
        keeperdict2[keepers1[i]] = gkseason2.loc[gkseason2['Player'] == keepers1[i], "ExpectedPSxG+/-"].item()
    else:
        continue
```

```python
firstseasongk = pd.DataFrame.from_dict(keeperdict1, orient = 'index')
firstseasongk = firstseasongk.reset_index()
firstseasongk.rename(columns = {'index': 'Name', 0: '17-18ExpectedPSxG+/-'}, inplace = True)
```

```python
secondseasongk = pd.DataFrame.from_dict(keeperdict2, orient = 'index')
secondseasongk = secondseasongk.reset_index()
secondseasongk.rename(columns = {'index': 'Name', 0: '18-19ExpectedPSxG+/-'}, inplace = True)
```

```python
keepersmerged = pd.merge(firstseasongk, secondseasongk, on = ["Name"])
```

```python
plt.rcParams["figure.figsize"] = (18, 5)
data_chart = keepersmerged.set_index('Name').plot.bar()
data_chart.set_xlabel('Goalkeeper/All Keepers ExpectedPSxG+/- Comparison Season 1 and 2')
data_chart.set_ylabel('ExpectedPSxG+/-')
data_chart.set_title("Performance of Huddersfield Town's goalkeepers 2017-18 VS 2018-19")
plt.show()
```

A common theme throughout our analysis of teams who had a particularly poor second season is that their goalkeepers had significantly worse second seasons then before. Huddersfield and Leeds United both had the same number 1 whereas Sheffield United had to change their goalkeeper due to Dean Henderson's loan expiring. Above we see that Jonas Lossl conceded 5 more goals than expected in 2017-18 but approximately 12 more in 2018-19.

### Defensive Actions

```python
defensivehudds1718 = pd.read_csv("datasets/cleanedAdvanced/huddersfield-defensive1718-cleaned.csv")
defensivehudds1819 = pd.read_csv("datasets/cleanedAdvanced/huddersfield-defensive1819-cleaned.csv")
```

```python
defensivehudds1718.tail()
```

```python
errors1 = list(defensivehudds1718['Player'])
errors2 = list(defensivehudds1819['Player'])
errorsdict1 = {}
errorsdict2 = {}
for i in range(len(errors1)):
    if errors1[i] in errors2:
        errorsdict1[errors1[i]] = defensivehudds1718.loc[defensivehudds1718['Player'] == errors1[i], "Errors"].item()
        errorsdict2[errors1[i]] = defensivehudds1819.loc[defensivehudds1819['Player'] == errors1[i], "Errors"].item()
    else:
        continue

firstseasonerrors = pd.DataFrame.from_dict(errorsdict1, orient = 'index')
firstseasonerrors = firstseasonerrors.reset_index()
firstseasonerrors.rename(columns = {'index': 'Name', 0: '17-18Errors'}, inplace = True)
secondseasonerrors = pd.DataFrame.from_dict(errorsdict2, orient = 'index')
secondseasonerrors = secondseasonerrors.reset_index()
secondseasonerrors.rename(columns = {'index': 'Name', 0: '18-19Errors'}, inplace = True)
```

```python
errorsmerged1 = pd.merge(firstseasonerrors, secondseasonerrors, on = ["Name"])
```

```python
errorsdf = errorsmerged1.iloc[:-1]
errorsdf.tail()
```

```python
plt.rcParams["figure.figsize"] = (18, 5)
#plt.style.use('fivethirtyeight') 
data_chart = errorsdf.set_index('Name').plot.bar()
data_chart.set_xlabel('Player Name')
data_chart.set_ylabel('Errors')
data_chart.set_title('Errors leading to shots 2017-18 VS 2018-19')
plt.show()
```

### Errors Analysis
It is also worth noting that teams who endure "second season syndrome" appear to make fewer mistakes leading to shots than in their first season and by a noticeable amount too. This trend has appeared across Huddersfield's, Sheffield United's and Leeds United's data and would be at odds at what you would expect to see. It suggests that these teams maybe got the "rub of the green" in some instances in their first season. With the PL a very competitive league, particularly around the middle to lower tier, not being punished by these errors could be the difference between getting points on the board or not.
